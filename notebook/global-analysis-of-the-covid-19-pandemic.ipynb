### **Handling Missing Value**



Mengatasi nilai null dalam dataset untuk memastikan integritas dan kualitas data yang lebih baik.

```python
# Menemukan baris yang mengandung nilai NaN

df_nan = df_filter[df_filter.isna().any(axis=1)]



# Menghitung jumlah baris yang mengandung nilai NaN

num_df_nan = df_nan.shape[0]



# Menampilkan jumlah baris yang mengandung nilai NaN

print(f"Jumlah baris yang mengandung nilai NaN: {num_df_nan}")
```

Berdasarkan hasil di atas terlihat bahwa terdapat banyak baris null pada dataset sehingga akan dilakukan visualisasi untuk melihat kolom apa saja yang memiliki nilai null paling banyak

Visualisasi Missing Value Setiap Kolom

```python
# Menghitung jumlah nilai NaN di setiap kolom

nan_counts = df_filter.isnull().sum()



# Barplot jumlah nilai NaN di setiap kolom

plt.figure(figsize=(15, 8))

nan_counts.plot(kind='bar', color='skyblue')

plt.title('Jumlah Nilai NaN di Setiap Kolom')

plt.xlabel('Kolom')

plt.ylabel('Jumlah Nilai NaN')

plt.show()
```

Dikarenakan hampir seluruh baris memiliki nilai NaN jadi Handling missing value yang dipilih adalah dengan mengisi nilai yang hilang menggunakan data sebelumnya (forward fill) karena dataset ini merupakan data time series. Handling missing value akan dilakukan hanya pada data bertipe numerik

```python
# Memisahkan kolom bertipe numerik dan non-numerik

numeric_cols = df_filter.select_dtypes(include='number').columns

non_numeric_cols = df_filter.select_dtypes(exclude='number').columns



# Menangani missing value hanya pada kolom numerik

df_numeric = df_filter[numeric_cols].copy()

df_numeric = df_filter.groupby('location')[numeric_cols].apply(lambda group: group.fillna(method='ffill'))

df_numeric = df_numeric.fillna(0)



# Menyatukan kembali kolom non-numerik terlebih dahulu, kemudian kolom numerik yang sudah diolah

df_notnull = pd.concat([df[non_numeric_cols].reset_index(drop=True), df_numeric.reset_index(drop=True)], axis=1)
```

```python
df_notnull.isnull().sum()
```

### **Data Transformation**

Mengubah tipe data kolom `date` menjadi datetime agar konsisten dan mudah digunakan dalam analisis data waktu.

```python
new_df = df_notnull.copy()



# Mengubah format kolom date

new_df['date'] = pd.to_datetime(new_df['date'])
```

Mengubah tipe data kolom numerik menjadi integer dengan membulatkan nilainya ke bilangan bulat, sehingga data lebih rapi dan mudah dianalisis secara efisien.

```python
# Membulatkan semua kolom numerik menjadi bilangan bulat

df_numeric = new_df.select_dtypes(include='number')  # Pilih kolom numerik

new_df[df_numeric.columns] = df_numeric.round(0).astype(int)  # Membulatkan dan mengonversi ke int



# Melihat informasi data

new_df.info()
```

## **EXPLANATORY DATA ANALYSIS (EDA)**




### **Analisis Global Kasus COVID-19 Tertinggi**



Melakukan analisis untuk menentukan negara dengan jumlah kasus Covid-19 tertinggi untuk mengidentifikasi wilayah yang paling terdampak

```python
# Mengurutkan data berdasarkan jumlah total kasus dan memilih 10 negara terbanyak

top_10_countries = df_clean.groupby('location')['total_cases'].max().nlargest(10)



# Mengonversi total kasus menjadi satuan juta

top_10_countries_million = top_10_countries / 1e6



# Mengatur gaya seaborn

sns.set(style="whitegrid")



# Membuat visualisasi dengan satu warna

plt.figure(figsize=(15, 8))

bar_plot = sns.barplot(x=top_10_countries_million.index, y=top_10_countries_million.values, color='steelblue')



# Menambahkan label di atas setiap bar

for index, value in enumerate(top_10_countries_million):

    bar_plot.annotate(f'{value:,.0f}', xy=(index, value), ha='center', va='bottom', fontsize=12)



# Mengatur label sumbu x

bar_plot.set_xticklabels(bar_plot.get_xticklabels(), rotation=45, horizontalalignment='center', fontsize=12)



# Mengatur label sumbu y dan x secara terpisah

bar_plot.set_ylabel('Total kasus (Juta)', fontsize=14)

bar_plot.set_xlabel('Negara', fontsize=14)



# Menambahkan judul

plt.title('10 Negara Dengan Jumlah Kasus Covid-19 Terbanyak', fontsize=16)



# Menghilangkan garis tepi

sns.despine(left=True, bottom=True)



# Menampilkan plot

plt.show()
```

Melakukan analisis untuk mengidentifikasi negara-negara dengan jumlah kasus COVID-19 tertinggi di seluruh dunia berdasarkan tahun.

```python
# Mengenkstrak tahun dari kolom date

df_clean['year'] = df_clean['date'].dt.year



# Fungsi untuk membuat visualisasi berdasarkan tahun

def plot_top_10_countries_by_year(df_clean, year, ax):

    df_year = df_clean[df_clean['year'] == year]

    top_10_countries = df_year.groupby('location')['total_cases'].max().nlargest(10)

    top_10_countries_million = top_10_countries / 1e6



    # Menggunakan satu warna (steelblue)

    sns.barplot(x=top_10_countries_million.index, y=top_10_countries_million.values, color='steelblue', ax=ax)



    # Menambahkan judul dan label sumbu

    ax.set_title(f'Top 10 Negara Dengan Total Kasus Terbanyak Tahun {year}', fontsize=16)

    ax.set_xlabel('Negara', fontsize=14)

    ax.set_ylabel('Total Kasus (Juta)', fontsize=14)

    ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='center', fontsize=12)



    # Menambahkan label di atas setiap bar

    for index, value in enumerate(top_10_countries_million):

        ax.annotate(f'{value:.0f}', xy=(index, value), ha='center', va='bottom', fontsize=12)



# Membuat subplot untuk tahun 2020, 2021, 2022, dan 2023

fig, axes = plt.subplots(5, 1, figsize=(16, 20))

years = [2020, 2021, 2022, 2023, 2024]



for i, year in enumerate(years):

    plot_top_10_countries_by_year(df_clean, year, axes[i])



plt.tight_layout()

plt.show()
```

**Insight**:



Kedua visualisasi di atas menunjukkan distribusi kasus COVID-19 berdasarkan jumlah total kasus secara global. Grafik pertama memperlihatkan 10 negara dengan total kasus COVID-19 tertinggi sepanjang pandemi, di mana Amerika Serikat dan China memiliki angka kasus yang jauh melampaui negara-negara lainnya. Grafik kedua membandingkan 10 negara dengan kasus tertinggi dari tahun 2020 hingga 2024. Dari grafik ini, terlihat bahwa Amerika Serikat secara konsisten berada di peringkat pertama sejak awal pandemi, dengan peningkatan kasus signifikan dari tahun ke tahun. Tren ini juga menunjukkan perkembangan kasus di negara lain seperti India, Brazil, dan China, yang mulai mengejar jumlah kasus tinggi pada tahun-tahun selanjutnya. Visualisasi ini memberikan wawasan tentang dinamika penyebaran pandemi di tingkat global selama periode waktu yang panjang.

### **Analisis Global Kasus Kematian Tertinggi**



Mengidentifikasi negara dengan angka kematian tertinggi akibat Covid-19 untuk menilai dampak fatal pandemi

```python
# Mengurutkan data berdasarkan jumlah total kasus dan memilih 10 negara terbanyak

top_10_countries = df_clean.groupby('location')['total_deaths'].max().nlargest(10)



# Mengonversi total kematian menjadi satuan juta

top_10_countries_million = top_10_countries / 1e3



# Mengatur gaya seaborn

sns.set(style="whitegrid")



# Membuat visualisasi

plt.figure(figsize=(15, 8))

bar_plot = sns.barplot(x=top_10_countries_million.index, y=top_10_countries_million.values, color='red')



# Menambahkan label di atas setiap bar

for index, value in enumerate(top_10_countries_million):

    bar_plot.annotate(f'{value:.0f}', xy=(index, value), ha='center', va='bottom')



# Menambahkan label pada sumbu

bar_plot.set_xticklabels(bar_plot.get_xticklabels(), rotation=45, horizontalalignment='center')

bar_plot.set(ylabel='Total Kematian (Ribu)', xlabel='Negara')

sns.despine(left=True, bottom=True)



# Menampilkan plot

plt.title("10 Negara Dengan Kasus Kematian Tertinggi (Ribu)")

plt.show()
```

```python
# Mengonversi kolom date menjadi datetime dan mengekstrak tahun

df_clean['year'] = df_clean['date'].dt.year



# Fungsi untuk membuat visualisasi berdasarkan tahun

def plot_top_10_countries_by_year(df_clean, year, ax):

    # Filter data untuk tahun tertentu

    df_year = df_clean[df_clean['year'] == year]



    # Menghitung 10 negara dengan total kematian tertinggi

    top_10_countries = df_year.groupby('location')['total_deaths'].max().nlargest(10)

    top_10_countries_million = top_10_countries / 1e3  # Konversi ke Ribu



    # Membuat bar plot

    sns.barplot(

        x=top_10_countries_million.index,

        y=top_10_countries_million.values,

        color='red',

        ax=ax

    )



    # Menambahkan judul dan label sumbu

    ax.set_title(f'Top 10 Negara Dengan Total Kematian Terbanyak Tahun {year}', fontsize=16)

    ax.set_xlabel('Negara', fontsize=14)

    ax.set_ylabel('Total Kematian (Ribu)', fontsize=14)

    ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='center', fontsize=12)



    # Menambahkan label di atas setiap bar

    for index, value in enumerate(top_10_countries_million):

        ax.annotate(f'{value:.0f}', xy=(index, value), ha='center', va='bottom', fontsize=12)



# Membuat subplot untuk tahun 2020, 2021, 2022, 2023, dan 2024

fig, axes = plt.subplots(5, 1, figsize=(16, 25))

years = [2020, 2021, 2022, 2023, 2024]



for i, year in enumerate(years):

    plot_top_10_countries_by_year(df_clean, year, axes[i])



# Menyesuaikan tata letak

plt.tight_layout()

plt.show()
```

Brasil, India, dan Rusia juga menduduki peringkat atas dengan jumlah kematian yang signifikan, mencerminkan dampak pandemi yang besar di negara-negara dengan populasi besar. Sementara itu, negara-negara seperti Meksiko, Peru, dan Inggris Raya menunjukkan stabilitas dalam jumlah kematian tahunan yang lebih rendah dibandingkan Amerika Serikat tetapi tetap berada dalam daftar 10 negara dengan kematian tertinggi. Secara keseluruhan, data ini menggambarkan dampak pandemi yang berbeda di setiap negara, dipengaruhi oleh berbagai faktor seperti populasi, respons kesehatan, dan tingkat vaksinasi.

### **Analisis tren jumlah kasus baru dan jumlah kematian baru**



Menganalisis perubahan jumlah kasus dan kematian baru untuk mengevaluasi tren penularan dan efektivitas intervensi kesehatan

```python
# Asumsi df_clean sudah ada dan memiliki kolom 'date', 'new_cases', dan 'new_deaths'

df_clean['date'] = pd.to_datetime(df_clean['date'])



# Mengambil bulan dan tahun dari kolom 'date'

df_clean['month_year'] = df_clean['date'].dt.to_period('M')



# Mengelompokkan data berdasarkan bulan dan tahun

monthly_data = df_clean.groupby('month_year')[['new_cases', 'new_deaths']].sum()



# Menghitung jumlah kasus baru dalam juta dan jumlah kematian baru dalam ribu

monthly_data['new_cases'] = monthly_data['new_cases'] / 1_000_000  # Menghitung dalam juta

monthly_data['new_deaths'] = monthly_data['new_deaths'] / 1_000  # Menghitung dalam ribu



# Mengatur tema seaborn

sns.set(style="whitegrid", palette="muted", font_scale=1.2)



# Membuat tampilan yang lebih besar

plt.figure(figsize=(18, 14))



# Plot untuk new_cases (Jumlah Kasus Baru dalam Jutaan)

plt.subplot(2, 1, 1)

sns.lineplot(x=monthly_data.index.astype(str), y=monthly_data['new_cases'], color='#1f77b4', lw=3, marker='o', markersize=8)

plt.title('Jumlah Kasus Baru Berdasarkan Bulan (dalam Juta)', fontsize=18, fontweight='bold')

plt.xlabel('Bulan', fontsize=14)

plt.ylabel('Jumlah Kasus Baru (dalam Juta)', fontsize=14)



# Mengatur interval sumbu x setiap 6 bulan dan menampilkan format bulan-tahun

plt.xticks(ticks=range(0, len(monthly_data), 6), labels=[month.strftime('%b-%Y') for month in monthly_data.index[::6]])

plt.xticks(rotation=45, ha='right')



# Plot untuk new_deaths (Jumlah Kematian Baru dalam Ribu)

plt.subplot(2, 1, 2)

sns.lineplot(x=monthly_data.index.astype(str), y=monthly_data['new_deaths'], color='#d62728', lw=3, marker='o', markersize=8)

plt.title('Jumlah Kematian Baru Berdasarkan Bulan (dalam Ribu)', fontsize=18, fontweight='bold')

plt.xlabel('Bulan', fontsize=14)

plt.ylabel('Jumlah Kematian Baru (dalam Ribu)', fontsize=14)



# Mengatur interval sumbu x setiap 6 bulan dan menampilkan format bulan-tahun

plt.xticks(ticks=range(0, len(monthly_data), 6), labels=[month.strftime('%b-%Y') for month in monthly_data.index[::6]])

plt.xticks(rotation=45, ha='right')



# Menambahkan teks untuk menonjolkan poin-poin penting

plt.tight_layout()

plt.show()
```

**Insight:**



**Jumlah Kasus Baru:**  

Jumlah kasus baru menunjukkan lonjakan signifikan pada awal 2022, mencapai puncak **95 juta** kasus per bulan, sebelum menurun tajam dan stabil di tingkat rendah mulai pertengahan 2023. Lonjakan kecil sempat terjadi awal 2023, namun tidak sebesar puncak sebelumnya. Tren ini mencerminkan pengendalian kasus yang membaik seiring waktu.



**Jumlah Kematian Baru:**  

Jumlah kematian baru mencapai puncaknya pada awal 2021, sebanyak **484 ribu** per bulan, lalu menurun secara bertahap dan stabil pada tingkat rendah sejak pertengahan 2022. Penurunan ini menunjukkan peningkatan efektivitas penanganan kesehatan, meskipun kasus baru sempat melonjak kembali pada 2023.

### **Analisis Total Kasus Covid-19 Berdasarkan Benua**



Membandingkan jumlah kasus Covid-19 di setiap benua untuk memahami pola tren penyebaran virus

```python
import matplotlib.dates as mdates



# Menambahkan kolom bulan dan tahun

df_clean['month'] = df_clean['date'].dt.to_period('M')



# Mengelompokkan data berdasarkan benua dan bulan

continent_trends = df_clean.groupby(['continent', 'month'])['new_cases'].sum().reset_index()



# Mengonversi kolom 'month' kembali ke datetime untuk visualisasi

continent_trends['month'] = continent_trends['month'].dt.to_timestamp()



# Menghitung jumlah kasus baru dalam juta

continent_trends['new_cases'] = continent_trends['new_cases'] / 1_000_000



# Membuat plot menggunakan seaborn lineplot

plt.figure(figsize=(12, 8))

sns.lineplot(data=continent_trends, x='month', y='new_cases', hue='continent')



# Menambahkan judul dan label

plt.title("Tren Perkembangan Kasus COVID-19 Berdasarkan Benua", fontsize=16)

plt.xlabel("Bulan", fontsize=12)

plt.ylabel("Jumlah Kasus Baru (dalam Juta)", fontsize=12)



# Mengatur format tanggal pada sumbu X

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b-%Y'))



# Menampilkan legend dan grid

plt.legend(title="Benua", fontsize=10)

plt.grid(True)

plt.tight_layout()



# Mengatur sumbu x

plt.xticks(rotation=45, ha='right', fontsize=10)



# Menampilkan grafik

plt.show()
```

**Insight:**



Grafik menunjukkan jumlah kasus COVID-19 baru di setiap benua dari 2020 hingga 2024. Eropa mencatat lonjakan terbesar pada awal 2022 dengan lebih dari **50 juta** kasus dalam sebulan, sementara Asia mengalami lonjakan tertinggi di awal 2023 dengan sekitar** 60 juta** kasus. Amerika Utara dan Amerika Selatan juga mengalami lonjakan pada periode yang sama, meskipun jumlahnya lebih kecil. Sebaliknya, Afrika dan Oseania mencatat jumlah kasus yang jauh lebih rendah dan stabil sepanjang periode.

Brazil dan Indonesia, meskipun memiliki populasi besar, menunjukkan angka vaksinasi booster yang lebih rendah, masing-masing 126 juta dan 69 juta, yang mungkin mencerminkan prioritas pada vaksinasi awal dibandingkan booster.
